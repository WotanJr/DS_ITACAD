Ens inclinem per no utilitzar el dropna. Omplirem tot el dataframe amb dades. Per aixo utilitzarem Imputation de la manera com ho hem comentat a la següent cel·la.

```python
# Utilitzem averages pels 8387 valors buits del retard a l'arribada, el temps a l'aire i el temps real de vol:
df_flights['ArrDelay'].fillna(df_flights['ArrDelay'].mean(), inplace = True)
df_flights['AirTime'].fillna(df_flights['AirTime'].mean(), inplace = True)
df_flights['ActualElapsedTime'].fillna(df_flights['ActualElapsedTime'].mean(), inplace = True)

# Alla on no tenim el temps de vol programat (CRSElapsedTime), utilitzem el valor del temps real (ActualElapsedTime):
df_flights['CRSElapsedTime'].fillna(df_flights['ActualElapsedTime'], inplace = True)

# Utilitzarem la moda per omplir les columnes TaxiIn i TaxiOut:
df_flights['TaxiIn'].fillna(df_flights['TaxiIn'].mode()[0], inplace = True)
df_flights['TaxiOut'].fillna(df_flights['TaxiOut'].mode()[0], inplace = True)

# Per l'hora d'arribada utilitzem la mediana:
df_flights['ArrTime'].fillna(df_flights['ArrTime'].median(), inplace = True)

# Utilitzarem zeros per omplir els temps de retard dels diferents conceptes. 
# Ho fem degut a que en tots aquests casos els retards estan per sota de 15 minuts per 
# tant considerem aquests retards menystenibles:
df_flights['CarrierDelay'].fillna(0, inplace = True)
df_flights['WeatherDelay'].fillna(0, inplace = True)
df_flights['NASDelay'].fillna(0, inplace = True)
df_flights['SecurityDelay'].fillna(0, inplace = True)
df_flights['LateAircraftDelay'].fillna(0, inplace = True)

# Omplirem amb "Unknown" les matricules dels avions que no quedaren registrades:
df_flights["TailNum"].fillna("Unknown", inplace = True)

print(df_flights.isna().sum())
```

Crearem 3 columnes: una amb la velocitat mitjana en milles per hora (AvgSpeed), una altra (Late) a on marquem si el vol te un retard de mes de 15 minuts i una altra (Date) amb la data a partir de les columnes de dia, mes i any.

```python
# Velocitat mitjana en milles per hora
df_flights_reduced["AvgSpeed"] = (df_flights_reduced["Distance"] * 60 / df_flights_reduced["ActualElapsedTime"]).round(2)
```

```python
# Ha tingut el vol un retard de més de 15 min? (1 = Si, 0 = no)
import numpy as np
df_flights_reduced["Late"] = np.where((df_flights_reduced['ArrDelay'] > 15), 1, 0)
```

```python
# Columna amb data YYYY-M-D
df_flights_reduced['Date'] = pd.to_datetime(dict(year=df_flights_reduced.Year, month=df_flights_reduced.Month, day=df_flights_reduced.DayofMonth))
```

Fes una taula de les aerolínies amb més endarreriments acumulats.

```python
def corr_coef(arr1, arr2):
    return(np.corrcoef(arr1, arr2))

print(corr_coef(df_flights_reduced["ArrDelay"], df_flights_reduced["DepDelay"]))
```

- Podem veure que hi ha una correlacio lineal positiva bastant forta entre el retard a la sortida i a l'arribada.
