## Quotes with BeautifulSoup

```python
#!python -m pip install requests
```

```python
import requests

URL_quotes = "http://quotes.toscrape.com/"
page_quotes = requests.get(URL_quotes)
page_quotes.raise_for_status
```

```python
#!pip install beautifulsoup4
```

```python
from bs4 import BeautifulSoup

soup_quotes = BeautifulSoup(page_quotes.content, "html.parser")

'''
Beautiful Soup object that takes page.content, which is the HTML content you scraped earlier, as its input.

Note: You'll want to pass page.content instead of page.text to avoid problems with character encoding. The .content attribute holds raw
bytes, which can be decoded better than the text representation you printed earlier using the .text attribute.

The second argument, "html.parser", makes sure that you use the appropriate parser for HTML content.
'''
```

```python
results_all = soup_quotes.find_all('div', class_="quote")
```

```python
for element in results_all:
    quote_element = element.find('span', class_="text")
    author_element = element.find('small', class_="author")
    print(quote_element.text)
    print(author_element.text, '\n')
```

## Quotes with Selenium

```python
#!pip install -U selenium
#!pip install webdriver-manager
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options


options = Options()
options.add_argument('--detach')

options
```

```python
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), 
                          options = options)
driver.get('http://quotes.toscrape.com//')

```

```python
from selenium.webdriver.common.by import By

texts = driver.find_elements(By.CSS_SELECTOR, '.text')
authors = driver.find_elements(By.CSS_SELECTOR, '.author')

for i in range(10):
    print(texts[i].text)
    print(authors[i].text, '\n')

driver.close()
```

## Cryptos chronology with Beautifulsoup

```python
import requests

URL_cryptos = "https://en.wikipedia.org/wiki/List_of_cryptocurrencies"
page_cryptos = requests.get(URL_cryptos)
page_cryptos.raise_for_status
```

```python
from bs4 import BeautifulSoup

soup_cryptos = BeautifulSoup(page_cryptos.content, "html.parser")
```

```python
tab = soup_cryptos.find('div', id="mw-content-text")

tabs = tab.find_all('table', class_="wikitable sortable")
```

```python
# sort through table
print("Criptomonedes (cronologia) \n")

for tt in tabs[0].find_all('tr')[1:]:        # skip first <tr> since it is the header
    print('Any: ', tt.find_all('td')[0].text + 'Nom : ',  tt.find_all('td')[1].text + 'Algoritme: ', tt.find_all('td')[4].text + 'Consensus: ', tt.find_all('td')[6].text.strip("[]") + 'Creadors: ', tt.find_all('td')[3].text)

    
```

## Cryptos chronology with Selenium

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options


options_c = Options()
options_c.add_argument('--detach')

options_c
```

```python
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

driver_c = webdriver.Chrome(service=Service(ChromeDriverManager().install()), 
                          options = options_c)
driver_c.get('https://en.wikipedia.org/wiki/List_of_cryptocurrencies')
```

```python
from selenium.webdriver.common.by import By

mytable = driver_c.find_element(By.XPATH, "//*[@id='mw-content-text']/div[1]/table[2]")   #pick the table

for row in mytable.find_elements(By.CSS_SELECTOR, 'tr'):     # sort through it
    i=0
    print('\n')
    for cell in row.find_elements(By.CSS_SELECTOR, 'td'):
        if i == 0:
            print('Any :', cell.text)
        if i==1:
            print('Coin :', cell.text)
        if i==4:
            print('Algoritme :', cell.text)
        if i==6:
            print('Consensus :', cell.text)
        if i==3:
            print('Creadors :', cell.text)
        i=i+1

driver_c.close()
```

## The Big Short with Selenium (film characters)

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

options = Options()
options.add_argument('--detach')

driver2 = webdriver.Chrome(service=Service(ChromeDriverManager().install()), 
                          options = options)
driver2.get('https://en.wikipedia.org/wiki/The_Big_Short_(film)')

mylist = driver2.find_element(By.XPATH, "//*[@id='mw-content-text']/div[1]/ul[1]")

actors = mylist.find_elements(By.CSS_SELECTOR, 'li')

for t in actors:
    print(t.text, '\n')

driver2.close()
```

## The Big Short with Selenium (book summary / characters)

```python

options = Options()
options.add_argument('--detach')

driver3 = webdriver.Chrome(service=Service(ChromeDriverManager().install()), 
                          options = options)
driver3.get('https://en.wikipedia.org/wiki/The_Big_Short')

myp1 = driver3.find_element(By.XPATH, "//*[@id='mw-content-text']/div[1]/p[3]")
myp2 = driver3.find_element(By.XPATH, "//*[@id='mw-content-text']/div[1]/p[4]")
myp3 = driver3.find_element(By.XPATH, "//*[@id='mw-content-text']/div[1]/p[5]")

print(myp1.text, '\n')
print(myp2.text, '\n')
print(myp3.text)

driver3.close()
```

Com a bonus farem un scaraping de la pagina de wikipedia d'una gran serie: Mr Robot. Mostrarem l'argument de la series i tots els actors.

```python

options = Options()
options.add_argument('--detach')

driver4 = webdriver.Chrome(service=Service(ChromeDriverManager().install()), 
                          options = options)
driver4.get('https://en.wikipedia.org/wiki/Mr._Robot')

mylist_r = driver4.find_element(By.XPATH, "//*[@id='mw-content-text']/div[1]/ul[1]")
mylist_rr = driver4.find_element(By.XPATH, "//*[@id='mw-content-text']/div[1]/ul[2]")

actors = mylist_r.find_elements(By.CSS_SELECTOR, 'li')
actors_rec = mylist_rr.find_elements(By.CSS_SELECTOR, 'li')


print('PLOT: \n')

print(driver4.find_element(By.XPATH, "//*[@id='mw-content-text']/div[1]/p[5]").text)

print('\nMAIN CAST: \n')
for t in actors:
    print(t.text, '\n')

    
print('RECURRING CAST: \n')
for a in actors_rec:
    print(a.text, '\n')

driver4.close()
```
